```python
with open("ceden_res_with_project.json", "r") as f:
    data = json.load(f)
```

```python
latitudes = [sample["Latitude"] for sample in data]
longitudes = [sample['Longitude'] for sample in data]
```

```python
coord_strings = [(sample["Longitude"],sample["Latitude"]) for sample in data]
```

```python
coordinates = pd.DataFrame({
   "Latitude": pd.Series(latitudes),
   "Longitude": pd.Series(longitudes),
   "coordinates": pd.Series(coord_strings)
})
coordinates 
```

```python
unique_coords = [(lat, lon) for lat, lon in coordinates["coordinates"].unique() if type(lat) == float or type(lat) == int]
```

```python
unique_coords
```

```python
lat, lon = unique_coords[0]
point = Point({"x": lat, "y": lon, "spatialReference": {"wkid": 4326}})

result = reverse_geocode(point)
address = result["address"]["Match_addr"]
print(address)
```

```python
cities = []
for lat, lon in unique_coords:
    coord_point = Point({"x": lat, "y": lon, "spatialReference": {"wkid": 4326}})
    result = reverse_geocode(coord_point)
    address = result["address"]["City"]
    cities.append(address)
```

```python
to_submit = pd.DataFrame({
    "Latitude": [lat for lat, lon in unique_coords],
    "Longitude": [lon for lat, lon in unique_coords],
    "City": cities
})
```

```python
to_submit.to_csv("coordinates_to_cities.csv")
```
