Load Libraries

```python
import numpy as np
import pandas as pd
import json
import csv
```

Load Data

```python
prism_data = pd.read_csv("Data/prism_4km.csv")
ceden_data = pd.read_csv("Data/ceden_data.csv")
prism_2022 = pd.read_csv("Data/prism_2022_4km.csv")
```

```python
ceden_data["coordinates"] = list(zip(ceden_data['Latitude'], ceden_data['Longitude']))
prism_2022["coordinates"] = list(zip(prism_2022['Latitude'], prism_2022['Longitude']))
```

```python
ceden_data["Date"] = pd.to_datetime(ceden_data["SampleDate"]).dt.date
prism_2022["Date"] = pd.to_datetime(prism_2022['Date']).dt.date
```

```python
ceden_data['Date'].max()
```

```python
merged_dataset_2022 = ceden_data.merge(prism_2022, on=['Date', "Latitude", "Longitude"], how='inner')
```

```python
merged_dataset_2022.to_csv("merged_prism_ceden_2022.csv")
```
